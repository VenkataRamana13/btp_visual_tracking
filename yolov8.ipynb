```python
import cv2
import numpy as np
from ultralytics import YOLO
import time

# Load the trained model
model = YOLO('/home/mtronlab/ramana/runs/obb/train13/weights/best.pt')

# Open the webcam (usually device 0)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open webcam")
    exit()

# Parameters for Lucas-Kanade Optical Flow
lk_params = dict(winSize=(15, 15), maxLevel=2,
                 criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

# Initialize variables
prev_gray = None
prev_points = None
tracking = False
trail_points = []  # List to store the tracked points and timestamps

# Duration to keep trail in seconds
trail_duration = 2

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # Convert frame to grayscale (needed for Optical Flow)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if not tracking:
        # Perform object detection
        results = model(frame)
        
        # Check if any objects were detected
        if results[0].boxes is not None and len(results[0].boxes.xyxy) > 0:
            # Get the first detected object (you can extend this for multi-object tracking)
            box = results[0].boxes.xyxy[0]  # First bounding box (x_min, y_min, x_max, y_max)

            # Get center of bounding box
            box = results[0].boxes.xyxy[0] 
            x_min, y_min, x_max, y_max = box
            center_x = (x_min + x_max) / 2
            center_y = (y_min + y_max) / 2

            # Initialize tracking points with the center of the detected object
            prev_points = np.array([[center_x, center_y]], dtype=np.float32).reshape(-1, 1, 2)
            prev_gray = gray.copy()  # Store the current grayscale image for tracking
            tracking = True  # Enable tracking

            # Add the initial point to the trail points list
            trail_points.append(((center_x, center_y), time.time()))

    else:
        # If tracking is enabled, use Optical Flow to track the object
        next_points, status, error = cv2.calcOpticalFlowPyrLK(prev_gray, gray, prev_points, None, **lk_params)

        # Update the previous frame and points
        prev_gray = gray.copy()
        prev_points = next_points

        # Extract the tracked point (center of the object)
        tracked_center = next_points[0].ravel()
        tracked_x, tracked_y = int(tracked_center[0]), int(tracked_center[1])

        # Add the tracked point to the trail points list with the current timestamp
        trail_points.append(((tracked_x, tracked_y), time.time()))

        # Remove points older than the trail duration
        current_time = time.time()
        trail_points = [(pt, t) for pt, t in trail_points if current_time - t < trail_duration]

        # Draw the fading trail
        for pt, t in trail_points:
            alpha = (current_time - t) / trail_duration  # Calculate alpha based on age of the point
            color_intensity = int((1 - alpha) * 255)  # Fade from full intensity to 0
            cv2.circle(frame, (int(pt[0]), int(pt[1])), 5, (0, 255, color_intensity), -1)  # Green trail with fading effect

        # Draw a circle around the current tracked object
        cv2.circle(frame, (tracked_x, tracked_y), 10, (0, 255, 0), -1)

        # Display tracking info
        cv2.putText(frame, f"Tracking Object at ({tracked_x}, {tracked_y})", 
                    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow('YOLO Object Detection + Optical Flow Tracking + Trail', frame)

    # Exit on 'q' key press
    if cv2.waitKey(1) == ord('q'):
        break

# When everything done, release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```

```python
import pyrealsense2 as rs
import numpy as np
import cv2
from ultralytics import YOLO

# Configure RealSense pipeline
pipeline = rs.pipeline()
config = rs.config()

# Enable depth and color streams
config.enable_stream(rs.stream.depth, 640, 480, rs.format.z16, 30)
config.enable_stream(rs.stream.color, 640, 480, rs.format.bgr8, 30)

# Start streaming
pipeline_profile = pipeline.start(config)

# Get depth sensor's depth scale
depth_sensor = pipeline_profile.get_device().first_depth_sensor()
depth_scale = depth_sensor.get_depth_scale()

# Align depth and color frames
align_to = rs.stream.color
align = rs.align(align_to)

# Load the YOLO model
model = YOLO('/home/mtronlab/ramana/runs/obb/train13/weights/best.pt')

# Set up video writer
frame_width = 640
frame_height = 480
out = cv2.VideoWriter('output.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width, frame_height))

try:
    while True:
        # Wait for a pair of aligned frames (depth + color)
        frames = pipeline.wait_for_frames()
        aligned_frames = align.process(frames)
        
        depth_frame = aligned_frames.get_depth_frame()
        color_frame = aligned_frames.get_color_frame()

        if not depth_frame or not color_frame:
            continue

        # Convert frames to numpy arrays
        depth_image = np.asanyarray(depth_frame.get_data())
        color_image = np.asanyarray(color_frame.get_data())

        # Perform object detection on color frame
        results = model(color_image)

        # Render the results on the frame
        annotated_frame = results[0].plot()  # Plot results on the frame

        if results[0].obb is not None and hasattr(results[0].obb, 'xyxy') and len(results[0].obb.xyxy) > 0:
            # Get the first bounding box
            x_min, y_min, x_max, y_max = results[0].obb.xyxy[0]

            # Calculate the center of the bounding box
            center_x = int((x_min + x_max) / 2)
            center_y = int((y_min + y_max) / 2)

            # Get the depth at the center of the bounding box
            depth = depth_frame.get_distance(center_x, center_y)
            depth_intrinsics = depth_frame.profile.as_video_stream_profile().intrinsics

            # Convert to 3D world coordinates
            point_3d = rs.rs2_deproject_pixel_to_point(depth_intrinsics, [center_x, center_y], depth)
#            text = f"Center: ({center_x}, {center_y}, {depth})"
            text = f"Center: ({point_3d})"

            # Display the center and depth information
            cv2.putText(annotated_frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Write the frame into the file 'output.avi'
        out.write(annotated_frame)

        # Display the resulting frame
        cv2.imshow('YOLO Object Detection', annotated_frame)

        # Exit on 'q' key press
        if cv2.waitKey(1) == ord('q'):
            break

finally:
    # Release everything after the loop
    pipeline.stop()
    out.release()
    cv2.destroyAllWindows()
```

```python
def pixel_to_world(u, v, Z, camera_matrix):
    # Intrinsic camera matrix (K)
    K_inv = np.linalg.inv(camera_matrix)

    # Pixel coordinates in homogeneous form
    pixel_coords = np.array([u, v, 1])

    # Calculate the world coordinates (X, Y, Z) where Z is the depth
    world_coords = Z * np.dot(K_inv, pixel_coords)

    X, Y = world_coords[:2]  # Extract X, Y from world_coords (Z is already known)
    return X, Y, Z
```

find intrinsics of camera using chessboard and opencv

```python
#!/usr/bin/env python
 
import cv2
import numpy as np
import os
import glob
 
# Defining the dimensions of checkerboard
CHECKERBOARD = (7,9)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
 
# Creating vector to store vectors of 3D points for each checkerboard image
objpoints = []
# Creating vector to store vectors of 2D points for each checkerboard image
imgpoints = [] 
 
 
# Defining the world coordinates for 3D points
objp = np.zeros((1, CHECKERBOARD[0] * CHECKERBOARD[1], 3), np.float32)
objp[0,:,:2] = np.mgrid[0:CHECKERBOARD[0], 0:CHECKERBOARD[1]].T.reshape(-1, 2)
prev_img_shape = None
 
# Extracting path of individual image stored in a given directory
images = glob.glob('./images/*.jpg')
for fname in images:
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    # Find the chess board corners
    # If desired number of corners are found in the image then ret = true
    ret, corners = cv2.findChessboardCorners(gray, CHECKERBOARD, cv2.CALIB_CB_ADAPTIVE_THRESH + cv2.CALIB_CB_FAST_CHECK + cv2.CALIB_CB_NORMALIZE_IMAGE)
     
    """
    If desired number of corner are detected,
    we refine the pixel coordinates and display 
    them on the images of checker board
    """
    if ret == True:
        objpoints.append(objp)
        # refining pixel coordinates for given 2d points.
        corners2 = cv2.cornerSubPix(gray, corners, (11,11),(-1,-1), criteria)
         
        imgpoints.append(corners2)
 
        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, CHECKERBOARD, corners2, ret)
     
    cv2.imshow('img',img)
    cv2.waitKey(0)
 
cv2.destroyAllWindows()
 
h,w = img.shape[:2]
 
"""
Performing camera calibration by 
passing the value of known 3D points (objpoints)
and corresponding pixel coordinates of the 
detected corners (imgpoints)
"""
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
 
print("Camera matrix : \n")
print(mtx)
print("dist : \n")
print(dist)
print("rvecs : \n")
print(rvecs)
print("tvecs : \n")
print(tvecs)
```
